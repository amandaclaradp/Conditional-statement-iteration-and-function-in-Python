## Problem 1

Tentukan dari variabel 
list_plat_mobil = ['B 1234 AB', 'B 6721 TY', 'B 1233 AY', 'B 6629 DD', 'B 1111 AM', 'B 6726 D', 'D 
11223 KJ', 'AE 44677 GH', 'AE 67269 AA']

Ada berapa jumlah plat genap dan ganjil?

```python
list_plat_mobil = ['B 1234 AB', 'B 6721 TY', 'B 1233 AY', 'B 6629 DD', 'B 1111 AM', 'B 6726 D', 'D 11223 KJ', 'AE 44677 GH', 'AE 67269 AA']
```

```python
# membuat fungsi untuk mendapatkan kode angka plat saja

def angka_plat(x):
  list_plat_angka = []
  for i in x:
    list_plat_angka.append(i.split()[1])
  return list_plat_angka
```

```python
# membuat fungsi untuk menentukan jenis plat

def jenis_plat(x):
  jenis = []

  for i in angka_plat(x):
    if int(i) % 2 == 0:
      jenis.append('genap')
    else:
      jenis.append('ganjil')
  return jenis
```

```python
# menghitung jumlah plat genap dan ganjil
jumlah_genap = jenis_plat(list_plat_mobil).count('genap')
jumlah_ganjil = jenis_plat(list_plat_mobil).count('ganjil')

print('Jumlah plat genap:', jumlah_genap)
print('Jumlah plat ganjil:', jumlah_ganjil)
```

## Problem 3

Team Campaign PT ABC akan membuat promo bundling category product (berisi 3 
category setiap bundling) untuk promo 17 Agustus. PT ABC memiliki 10 category product. 
Ketentuan Bundling:
1. Tidak boleh hanya memiliki 1 jenis category dalam 1 bundling
2. 1 Bundling boleh memiliki 2 category yang sama
3. Urutan tidak menjadi faktor pembeda (c1 - c2 - c3 = c2 - c1 - c3).

Tentukan ada berapa jumlah kemungkinan bundling yang bisa terbentuk?

```python
list_id_category = ['c1', 'c2', 'c3', 'c4', 'c5', 'c6', 'c7' , 'c8', 'c9', 'c10']
jumlah_bundling = 0

for i in range(len(list_id_category)):
    for j in range(i, len(list_id_category)):
        for k in range(j, len(list_id_category)):
            if i != j and j != k or i != k:
              jumlah_bundling += 1

print("Jumlah kemungkinan bundling:", jumlah_bundling)
```

## Problem 4

Buatlah sebuah function string_processor dengan inputan 2 parameter, yaitu suatu 
string 1 kata dan nama proses yang akan dilakukan. Function ini memiliki 4 proses
- first_char : return huruf pertama dari string yang diinput
- last_char : return huruf terakhir dari string yang diinput
- num_chars : return jumlah huruf string input
- num_vowels : return jumlah huruf vocal pada string input

```python
def string_processor(kata, proses):
  if proses.lower() == 'first_char':
    kata = kata[0]
    return kata
  elif proses.lower() == 'last_char':
    kata = kata[-1]
    return kata
  elif proses.lower() == 'num_chars':
    kata = len(kata)
    return kata
  elif proses.lower() == 'num_vowels':
    count = 0
    for i in kata:
        if i in ['a','i','u','e','o']:
            count += 1
    return count
```

```python
# memanggil fungsi string_processor
string_processor('indah', proses = 'last_char')
```

## Problem 5

Menggunakan dataset rakamin_store_city_monthly! Buatlah 1 kolom baru dengan nama kolom is_gmail 
dengan value True jika menggunakan gmail dan False jika selain gmail. 

Tentukan ada berapa user yang menggunakan email gmail pada tahun 2018!

```python
# Read data

df = pd.read_csv("rakamin_store_city_monthly.csv", sep=';')

df.head()
```

```python
# membuat kolom tahun
df['Year'] = df.apply(lambda x: x['month'][:4], axis=1)
df.head()
```

```python
# menambahkan kolom is_gmail
df['is_gmail'] = df['email'].apply(lambda x: True if x.split('@')[1] == 'gmail.com' else False)
df.head()
```

```python
# menghitung user yang menggunakan gmail pada tahun 2018
df_filter = df[(df['Year'] == '2018') & (df['is_gmail'] == True)]

jumlah_user = df_filter['user_id'].count()
print('Jumlah user yang menggunakan email gmail pada tahun 2018 adalah', jumlah_user)
```

## Problem 6

Menggunakan dataset superstore! Buatlah function yang bisa menghitung Jumlah Sales 
dan Profit pada suatu tahun dan region tertentu dengan default tahun 2017. 
Tentukan total sales dan profit:
1. South Region pada tahun 2016
2. West Region pada tahun 2017

```python
# Read data

data = pd.read_excel("superstore.xls")

data.head()
```

```python
# membuat kolom tahun
data['Year'] = data['Order Date'].dt.year

data.head()
```

```python
# membuat funsi untuk menghitung jumlah sales dan profit
def total_order(region, year=2017):
  data_user_selected = data.groupby(['Region', 'Year']).agg({'Sales' : 'sum',
                                                          'Profit' : 'sum'}).reset_index()
  
  total_sales_profit = data_user_selected[(data_user_selected['Region'] == region) & (data_user_selected['Year'] == year)]
  total_sales = total_sales_profit['Sales'].iloc[0]
  total_profit = total_sales_profit['Profit'].iloc[0]
  
  print('Total sales pada', region, 'region tahun', year, 'adalah', total_sales, 'dengan total profit', total_profit)
```

```python
# mendapatkan total sales dan profit untuk south region tahun 2016
total_order('South', year=2016)
```

```python
# mendapatkan total sales dan profit untuk west region tahun 2017
total_order('West')
```

## Problem 7

Menggunakan dataset superstore! Buatlah 3 kelompok 'segment' (high, medium, dan 
low) berdasarkan performance Sales yang digunakan sebagai acuan tim bisnis untuk 
mengelompokan customer superstore.

```python
# agregasi user

data_agg = data.groupby(['Customer ID', 'Year']).agg({'Sales' : 'sum'}).reset_index()

data_agg.columns = ['Customer ID', 'Year', 'Total Sales']

data_agg.head()
```

```python
# mendapatkan data statistik dari data yang sudah diagregasi di tingkat user

data_agg.describe()
```

Untuk menentukan treshold tiap segmen, data akan dibagi menjadi 3 sama besar dengan treshold pada percentile 33,33 dan percentile 66,67

```python
# mendapatkan percentile 33,33 dan 66,67
data_agg.describe(percentiles=[.3333,.6667])
```

Berdasarkan hasil di atas dapat ditentukan batasan total sales untuk setiap segment sebagai berikut:
- Low jika total sales <= 208.983276
- Medium jika 208.983276 < total sales <= 864.998276
- High jika total sales > 864.998276

```python
# membuat function segment
def segment(x):
  try:
    if (x['Total Sales'] <= 208.983276) and (x['Total Sales'] >= 0):
      segment = 'Low'
    elif (x['Total Sales'] > 208.983276) and (x['Total Sales'] <= 864.998276):
      segment = 'Medium'
    elif x['Total Sales'] > 864.998276:
      segment = 'High'
  except:
    segment = 'Unidentified'
  return segment
```

```python
# menambahkan kolom segment
data_agg['Segmentasi'] = data_agg.apply(lambda x : segment(x), axis = 1)
data_agg.head()
```

```python
# mendapatkan jumlah customer masing-masing segment tahun 2017
total_customer = data_agg[data_agg['Year'] == 2017].groupby('Segmentasi').agg({'Customer ID' : 'count'}).reset_index()
total_customer.columns = ['Segmentasi', 'Jumlah Customer']
total_customer
```
